```python
class ShoppingCart:
    def __init__(self, customer_name, items=None):
        self.customer_name = customer_name
        self.items = items or []


    """Add a price (must be positive number)"""
    def add_item(self, item):
        if item <= 0:
            print("\n")
            print(f"Invalid price! '{item}' is not a valid positive number.")
            print("_"*60)
        else:
            self.items.append(item)
            print("\n")
            print(f"{item} added to the cart successfully.")
            print("_"*60)


    """Remove an item by value"""
    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
            print("\n")
            print(f"{item} removed from the cart.")
            print("_"*60)
        else:
            print("\n")
            print(f"{item} not found in the cart.")
            print("_"*60)


    """Return sum of all items"""
    def calculate_total(self):
        total = 0.0
        for item in self.items:
            total += item
        return total


    """Apply 10% discount if total >= 3000"""
    def apply_discount(self):
        total = self.calculate_total()
        if total >= 3000:
            discount_amount = total * 0.10
            return total - discount_amount
        else:
            return total


    """Cart display"""
    def display_cart(self):
        print("\n")
        print("="*60)
        print(f"SHOPPING CART - {self.customer_name.upper()}")
        print("="*60)

        if not self.items:
            print("Your cart is currently empty.")
        else:
            print(f"Items in cart: {len(self.items)}")
            print("-" * 60)
            for i, item in enumerate(self.items, 1):
                print(f"   {i:2d}. {item:.2f}")

        total = self.calculate_total()
        discounted_total = self.apply_discount()

        print("-" * 60)
        print(f"   Total             : {total:8.2f}")


        discount = total * 0.10 if total >= 3000 else 0.0
        print(f"   Discount(-)       : {discount:8.2f}")
        print(f"   Final Total       : {discounted_total:8.2f}")


        print("="*60)


    """Clear the entire cart"""
    def clear_cart(self):
        self.items.clear()
        print("\n")
        print(f"Cart for {self.customer_name} has been cleared successfully.")
        print("_"*60)


    """Combine two carts using + operator"""
    def __add__(self, other):
        if self.customer_name == other.customer_name:
            new_name = self.customer_name
        else:
            new_name = f"{self.customer_name} & {other.customer_name}"

        combined_items = self.items + other.items
        return ShoppingCart(new_name, combined_items)
```

## Part A – Class Creation

```python
customer1 = ShoppingCart("Jubayer", [1000,2000])
customer2 = ShoppingCart("Shakib")

print(f"Customer 1 name: {customer1.customer_name}")
print(f"Customer 2 name: {customer2.customer_name}")

print(f"Customer 1 items: {customer1.items}")
print(f"Customer 2 items: {customer2.items}")
```

## Part B – Core Methods Implementation

```python
customer2.add_item(1000)
customer2.add_item(2000)
customer2.add_item(-3000)

customer1.remove_item(1000)
customer1.remove_item(3000)
```

```python
print(f"Customer 1 total: {customer1.calculate_total()}")
print(f"Customer 2 total: {customer2.calculate_total()}")
```

```python
print(f"Customer 1 discounted total: {customer1.apply_discount()}")
print(f"Customer 2 discounted total: {customer2.apply_discount()}")
```
